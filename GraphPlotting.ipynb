```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV file
# Replace 'your_data.csv' with the path to your CSV file
data = pd.read_csv('STNV_WT_25ms_local_reac.csv')

# Assume your CSV has columns named 'x', 'y', and 'se' for the x values, y values, and standard errors
x = data['genome_position']
y = data['reactivity']
se = data['se_reac']

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=se, fmt='', 
             #capsize=5, these were the caps on the error bars  I have removed them for this graph 
             #label='Data with SE', this was the legend that I don't need here
             color='blue', linestyle='-', linewidth=2, ecolor='red')

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('VLP_Reactivity')
#plt.legend()
plt.grid()

# Save the plot
plt.savefig('plot.eps', format='eps', dpi=300)  # Save as PNG with 300 dpi

# Show the plot
plt.show()
```
